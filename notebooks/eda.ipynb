```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
# Get the absolute path of the project root (one level up from 'notebooks')
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))

# Construct the correct file path
file_path = os.path.join(project_root, "data", "processed_data.parquet")

# Load the file
df = pd.read_parquet(file_path)

print("Data loaded successfully!")
```

```python
# Basic statistics
print(df.describe())
```

```python
# Visualize distribution of "Total Time Spent"
sns.histplot(df["Total Time Spent"], bins=30, kde=True)
plt.title("Distribution of Total Time Spent")
plt.show()
```
